```python
# Returns the DataFrame of the Engineered Features
def featureEngineering(filepath):
    # First Load Data
    df = pd.read_csv(filepath, encoding="ISO-8859-1")
    
    # Now set up x, y for data and results
    features = pd.DataFrame()
    labels = (df["type"] == "spam").astype(int)

    # Store texts to be reused
    textList = df["text"] 

    # Set up the Feature Engineered Columns
    features["length"] = numChar(textList)
    
    # Training & Testing Split
    return features, labels
```
